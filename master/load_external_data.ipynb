```python
def find_in_rigla(keyword):
    url = 'http://www.rigla.ru/search/?q=' + str(keyword.replace(' ', '+'))
    headers = {
                'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.9; rv:45.0) Gecko/20100101 Firefox/45.0'
              }

    r = requests.get(url, headers = headers)

    soup = BeautifulSoup(r.text.encode('utf8'), "lxml")

    prices = []
    for price in soup.findAll('span', {'class': 'price'}):
        prices.append([int(s) for s in price.text.split() if s.isdigit()][0])
    if len(prices) == 1: 
        return prices[0]
    elif len(prices) > 1:
        return str(min(prices)) + ' - ' + str(max(prices))
```
